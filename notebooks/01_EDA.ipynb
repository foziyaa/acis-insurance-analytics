```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set some display options for better viewing in the notebook
pd.set_option('display.max_columns', 100)
pd.set_option('display.width', 1000) # Helps when printing df.head() in some terminals

# --- The Updated Part ---

# 1. Define the correct file path
file_path = '../data/raw/MachineLearningRating_v3.txt'

try:
    # 2. Use pd.read_csv but specify the pipe delimiter
    df = pd.read_csv(file_path, delimiter='|')
    
    print("Data loaded successfully! Here are the first 5 rows:")
    display(df.head())

except FileNotFoundError:
    print(f"ERROR: The file was not found at {file_path}")
    print("Please make sure you have downloaded the data and placed it in the 'data/raw/' directory.")
except Exception as e:
    print(f"An error occurred while loading the data: {e}")
```

```python
# Check data types and non-null counts
df.info()

# Get statistical summary for numerical columns
df.describe()
```

```python
# Check for missing values
df.isnull().sum().sort_values(ascending=False)
```

```python
# See the count of nulls in each column, sorted high to low
df.isnull().sum().sort_values(ascending=False)
```

```python
# Set up plotting style
sns.set_style('whitegrid')

# Plot a histogram of TotalPremium
plt.figure(figsize=(10, 6))
sns.histplot(df['TotalPremium'], bins=50, kde=True)
plt.title('Distribution of Total Premium')
plt.xlabel('Total Premium')
plt.ylabel('Frequency')
plt.show()

# Do the same for TotalClaims, but be careful of the range
# Since many claims are 0, let's look at claims > 0
plt.figure(figsize=(10, 6))
sns.histplot(df[df['TotalClaims'] > 0]['TotalClaims'], bins=50, kde=True)
plt.title('Distribution of Claim Amounts (for policies with claims)')
plt.xlabel('Total Claims')
plt.ylabel('Frequency')
plt.show()
```

```python
# --- Bivariate Analysis: Loss Ratio by Province ---

# First, let's calculate the overall Loss Ratio for the entire portfolio
total_claims = df['TotalClaims'].sum()
total_premium = df['TotalPremium'].sum()
overall_loss_ratio = total_claims / total_premium

print(f"Overall Portfolio Loss Ratio: {overall_loss_ratio:.2%}\n")

# Now, let's calculate it for each province
# We group by Province, sum up the claims and premiums for each, and then calculate the ratio
province_financials = df.groupby('Province')[['TotalClaims', 'TotalPremium']].sum()

# Avoid division by zero if a province has no premium
province_financials['LossRatio'] = province_financials['TotalClaims'] / province_financials['TotalPremium']

# Sort the results to easily see the best and worst provinces
province_financials_sorted = province_financials.sort_values(by='LossRatio', ascending=False)

print("Loss Ratio by Province:")
print(province_financials_sorted)

# --- Visualization ---

plt.figure(figsize=(12, 7))
sns.barplot(x=province_financials_sorted.index, y=province_financials_sorted['LossRatio'], palette='viridis')

# Add a horizontal line for the overall average to see which provinces are above/below average
plt.axhline(y=overall_loss_ratio, color='r', linestyle='--', label=f'Overall Avg ({overall_loss_ratio:.2%})')

plt.title('Loss Ratio by Province', fontsize=16)
plt.xlabel('Province', fontsize=12)
plt.ylabel('Loss Ratio (Claims / Premium)', fontsize=12)
plt.xticks(rotation=45, ha='right') # Rotate labels for better readability
plt.legend()
plt.tight_layout() # Adjust plot to prevent labels from being cut off
plt.show()
```

```python
# --- Outlier Detection ---

# Create a figure with two subplots side-by-side
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Box plot for Total Premium
sns.boxplot(y=df['TotalPremium'], ax=axes[0])
axes[0].set_title('Box Plot of Total Premium')
axes[0].set_ylabel('Premium Amount')

# Box plot for Total Claims (only for policies that had a claim)
# This gives a more meaningful view of claim amounts
sns.boxplot(y=df[df['TotalClaims'] > 0]['TotalClaims'], ax=axes[1])
axes[1].set_title('Box Plot of Total Claims (Claims > 0)')
axes[1].set_ylabel('Claim Amount')

plt.tight_layout()
plt.show()
```

```python
# --- Bivariate Analysis: Top 10 Riskiest Vehicle Makes by Loss Ratio (CORRECTED) ---

# We only want to look at makes with a reasonable number of policies
# to avoid skewed results from rare makes. Let's set a threshold.
# Using the correct lowercase 'make' column
policy_counts = df['make'].value_counts()
common_makes = policy_counts[policy_counts > 50].index # Only consider makes with >50 policies

df_common_makes = df[df['make'].isin(common_makes)]

# Now calculate loss ratio for these common makes
# Using the correct lowercase 'make' column
make_financials = df_common_makes.groupby('make')[['TotalClaims', 'TotalPremium']].sum()
make_financials['LossRatio'] = make_financials['TotalClaims'] / make_financials['TotalPremium']

# Get the top 10 riskiest
top_10_risky_makes = make_financials.sort_values(by='LossRatio', ascending=False).head(10)

print("Top 10 Riskiest Vehicle Makes (by Loss Ratio):")
print(top_10_risky_makes)

# --- Visualization ---
plt.figure(figsize=(12, 7))
sns.barplot(x=top_10_risky_makes.index, y=top_10_risky_makes['LossRatio'], palette='rocket')
# Let's re-calculate overall_loss_ratio here just to be safe it's in scope
overall_loss_ratio = df['TotalClaims'].sum() / df['TotalPremium'].sum()
plt.axhline(y=overall_loss_ratio, color='b', linestyle='--', label=f'Overall Avg ({overall_loss_ratio:.2%})')

plt.title('Top 10 Riskiest Vehicle Makes by Loss Ratio', fontsize=16)
plt.xlabel('Vehicle Make', fontsize=12)
plt.ylabel('Loss Ratio', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```
